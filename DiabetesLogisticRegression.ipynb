```python
## Import librari sklearn untuk membagi antara variabel menggunakan "train_test_split"
## Import librari sklearn untuk mengecek score machine learning menggunakan "accuracy_score"

from sklearn import linear_model
from sklearn.metrics import mean_absolute_error as mae,median_absolute_error as mee,classification_report as cr,accuracy_score as ac
```

**PEMBAGIAN VARIABEL DEPENDENT DAN INDPENEDENT**

```python
## Mendefinisikan Variabel Dependent dan Variabel Independent
feature_columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age' ]
predicted_class = ['Outcome'] 
```

```python
## Mendefinisikan variabel Independen as X, Variabel dependent as Y
X = db[feature_columns].values
Y = db[predicted_class].values
```

**DIVIDE THE VARIABEL FOR TRAINING AND TESTING FOR MACHINE LEARNING**

```python
## Membagi training dan testing variabel dengan ukuran dari testingnya sebesar 30%
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30, random_state=10)
```

**PEMBUATAN MODEL LOGISTIC REGRESSION DAN TRAIN MACHINE**

```python
## Import Module Logistic Regression
from sklearn.linear_model import LogisticRegression

## Inisialisasi Model Logisticregression
logistic_regression = LogisticRegression()
```

```python
## Training model LogicticRegression 
logistic_regression.fit(X_train,Y_train)

## Melakukan prediksi pada set pengujian
y_prediction = logistic_regression.predict(X_test)
```

**CEK KINERJA MODEL MELALUI CONFUSION MATRIKS**

```python
## import Module metrics dari Sklearn
from sklearn import metrics

## Menentukan kinerja model dengan confusion matrix
confusion_matrix_outcome = metrics.confusion_matrix(Y_test, y_prediction)
confusion_matrix_outcome
```

```python
# numpy, matplotlib.pyplot, seaborn are imported
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**MEMVISUALISASIKAN CONFUSION MATRIKS**

```python
# name  of classes is being represented by names_detail
names_detail=[0,1] 
fig, ax = plt.subplots()
plt.xticks(np.arange(len(names_detail)), names_detail)
plt.yticks(np.arange(len(names_detail)), names_detail)

# plotting the heatmap as per the confusion matrix (confusion_matrix_outcome)
sns.heatmap(pd.DataFrame(confusion_matrix_outcome), annot=True, cmap="Blues" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.title('Confusion matrix for model evaluation', y=1.5)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

**PENGECEKAN AKURASI MODEL**

```python
## kalkulasi akurasi dari model
print("Accuracy of the model is:",metrics.accuracy_score(Y_test, y_prediction))
```
