```python
import pandas as pd
import numpy as np
```

```python
excel = pd.read_csv("heart_2020_cleaned.csv")
excel
```

```python
dt = excel.corr()
dt.style.background_gradient(cmap='coolwarm')
```

```python
x = excel.drop(["HeartDisease","Smoking","AlcoholDrinking","Stroke","DiffWalking",
                "Sex","AgeCategory","Race","Diabetic","PhysicalActivity","GenHealth","Asthma",
                "KidneyDisease","SkinCancer"], axis=1)
x.head(5)

##jvkhbjk
```

```python
y = excel['HeartDisease']
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
nbc = GaussianNB()
```

```python
data_training = nbc.fit(x_train, y_train)
```

```python
y_predict = data_training.predict(x_test)
print(y_predict)
```

```python
## Example people
BMI = 27.41
PhysicalHealth = 7.0
MentalHealth = 0.0
SleepTime = 6.0

##testing
Data_Testing = [[BMI,PhysicalHealth,MentalHealth,SleepTime]]
```

```python
y_pred = data_training.predict(Data_Testing)
```

```python
if y_pred == 'No':
    hasil = "Tidak terkena"
elif y_pred == 'Yes':
    hasil = "Akan terkena"
else:
    hasil = "eror"
    
print("Resiko Terkena Penyakit Jantung =",hasil)
```

```python
from sklearn.metrics import accuracy_score
print("nilai akurasi = %0.2f " % accuracy_score(y_test,y_predict))
```

```python
score = accuracy_score(y_test, y_predict)
print ('Nilai akurasi :', float(score))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y,y_predict))
```
